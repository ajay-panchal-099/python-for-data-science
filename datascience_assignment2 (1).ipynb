```python
# ASSIGNMENT 1
# TASK - 1
# question 1.1

"""Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()"""

def myreduce(func,collection):
    result = collection[0]
    for i in range(1,len(collection)):
        result = func(result,collection[i])
    return result

def func(a,b):
    return a+b;

l = [1,2,3,4,5]
print("this is myreduce function: ",myreduce(func,l))
        
```

```python
# question 1.2
"""Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()"""

def myfilter(func1,collection):
    result=[]
    for j in collection:
        if(func1(j)):
            result.append(j)
    return result

def func1(a):
    if(a>0):
        return True
    else:
        return False
    
ll = [1,2,3,-1,-4,-6,13,-24]
print("this is my filter function: ",myfilter(func1,ll))   
    
```

```python
# TASK -2 
# QUESTION - 1.1

"""Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass."""


class triangle:
    
    def __init__(self,f,se,t):
        self.a=f
        self.b=se
        self.c=t
    
class compute_area(triangle):
    
    def __init__(self,A,B,C):
        super().__init__(A,B,C)
        
    def area(self):
        s=(self.a+self.b+self.c)/2
        return (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5

f=int(input("Enter The first side length of triangle--> "))
se=int(input("Enter The second side length of triangle --> "))
t=int(input("Enter The third side length of triangle --> "))
p=compute_area(f,se,t)
p.area()
```

```python
# TASK -2
# QUESTION - 1.2

"""Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n."""

def filter_long_words(L,N):
    for i in L:
        if len(i)>N:
            m.append(i)
    return m


L=[]
m=[]
N=int(input("Enter The Value N : "))
p=int(input("Enter The List Length : "))
print("Enters The Words ")
for i in range(p):
    s=input(" ")
    L.append(s)
    
print("Words Whose Lenght is greater Then ",N," are :    ",filter_long_words(L,N))   

```

```python
# TASK - 2
# QUESTION 2.2
"""Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is
a vowel, False otherwise. """


t=['a','e','i','o','u','A','E','I','O','U']
def check(s):
    if s in t:
        return True
    else:
        return False
s=input("Enter A Character : ")
print("The Entered Character is Vowel :", check(s))
```
